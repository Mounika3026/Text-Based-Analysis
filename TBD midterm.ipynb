# stop words

```python
import re

# Define a list of common English stop words

stop_words = ["the", "is", "in", "at", "which", "on", "and", "or", "but", "an", "a", "to", "for"]

# Create a regex pattern for stop words

stop_words_pattern = r'\b(' + '|'.join(stop_words) + r')\b'

# Remove stop words from a specific text column (assuming the text column is 'Review Text')

df['Review Text No Stopwords'] = df['Review Text'].apply(
    lambda x: re.sub(stop_words_pattern, '', x, flags=re.IGNORECASE)
)

# Clean extra spaces caused by removing stop words

df['Review Text No Stopwords'] = df['Review Text No Stopwords'].str.replace(r'\s+', ' ', regex=True).str.strip()

# Display the first 20 rows to see the changes

df.head(20)
```

```python
## Function to tokenize:

import regex as re
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

```python
## make a pipeline

pipeline = [str.lower, tokenize]

def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

```python
## Run the pipeline

df['tokens'] = df['Review Text'].apply(prepare, pipeline=pipeline)
```

```python
## Count the tokens

df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

```python
## Use the Counter

from collections import Counter

from collections import Counter


# Function to count word frequency

def word_frequency(tokens):
    return Counter([token.lower() for token in tokens])


# Perform word frequency analysis on the tokens without stop words

word_freq = word_frequency("Review Text No Stopwords")

print("Word Frequency:", word_freq)
```

```python
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

```python
## word frequency in its owndataframe for analysis

freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

```python
## sort and display

freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

```python
from textblob import TextBlob
from wordcloud import WordCloud, STOPWORDS
```

```python
#Create a function to get the subjectivity
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

#Create a function to get Polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity

#Now we create a new column for what we just did and add it to the Tweet_df dataframe
df['Subjectivity'] = df['Review Text'].apply(getSubjectivity)
df['Polarity'] = df['Review Text'].apply(getPolarity)

#Now display data
df.head()
```

```python
#Group the range of Polarity to different categories
def getInsight(score):
    if score < 0:
        return "Negative"
    elif score == 0:
        return "Neutral"
    else:
        return "Positive"
    
df["Insight"] = df["Polarity"].apply(getInsight)

df.head(50)
```

```python
import seaborn as sns
import warnings
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

# 5 random strings

```python
import re

def test_regex_patterns():
    # 1. Email Addresses
    email_pattern = r'^[\w.%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'
    email_example = "mounika123@example.com"
    is_email = re.match(email_pattern, email_example)
    print("Email:", email_example, "Valid:", bool(is_email))
    
    # 2. File Paths (Unix-based systems)
    file_path_pattern = r'^/([a-zA-Z0-9_-]+/)*([a-zA-Z0-9_-]+\.[a-zA-Z0-9_-]+)$'
    file_path_example = "/home/mounika/documents/file.txt"
    is_file_path = re.match(file_path_pattern, file_path_example)
    print("File Path:", file_path_example, "Valid:", bool(is_file_path))
    
    # 3. IP Addresses (IPv4)
    ip_pattern = r'^(25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.(25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.(25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.(25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)$'
    ip_example = "192.168.1.1"
    is_ip = re.match(ip_pattern, ip_example)
    print("IP Address:", ip_example, "Valid:", bool(is_ip))
    
    # 4. Phone Numbers (US format)
    phone_pattern = r'^\(\d{3}\) \d{3}-\d{4}$'
    phone_example = "(123) 456-7890"
    is_phone = re.match(phone_pattern, phone_example)
    print("Phone Number:", phone_example, "Valid:", bool(is_phone))
    
    # 5. Dates (ISO 8601)
    date_pattern = r'^\d{4}-\d{2}-\d{2}$'
    date_example = "2024-04-27"
    is_date = re.match(date_pattern, date_example)
    print("Date:", date_example, "Valid:", bool(is_date))

# Call the function to test the regex patterns
test_regex_patterns()
```
